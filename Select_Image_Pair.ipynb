```python
import cv2

# Path to the video file
video_path = 'output.avi'

# Open the video
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Get the frame rate of the video
fps = cap.get(cv2.CAP_PROP_FPS)

# Calculate the frame indices for 2 seconds apart
frame_index1 = int(fps * 4)  # Frame at 2 seconds
frame_index2 = int(fps * 7)  # Frame at 4 seconds (2 seconds apart from the first frame)

# Read and extract the first frame
cap.set(cv2.CAP_PROP_POS_FRAMES, frame_index1)
ret, frame1 = cap.read()
if ret:
    cv2.imwrite('image1.jpg', frame1)
    print("Image 1 saved successfully.")
else:
    print("Error: Failed to capture Image 1.")

# Read and extract the second frame
cap.set(cv2.CAP_PROP_POS_FRAMES, frame_index2)
ret, frame2 = cap.read()
if ret:
    cv2.imwrite('image2.jpg', frame2)
    print("Image 2 saved successfully.")
else:
    print("Error: Failed to capture Image 2.")

# Release the video capture object
cap.release()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the images
image1 = cv2.imread('image1.jpg')
image2 = cv2.imread('image2.jpg')

# Convert images from BGR to RGB
image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

# Create a figure to display the images
plt.figure(figsize=(10, 5))  # Set the figure size

# Display Image 1
plt.subplot(1, 2, 1)  # (rows, columns, panel number)
plt.imshow(image1_rgb)
plt.title('Frame 1')
plt.axis('off')  # Hide axes ticks

# Display Image 2
plt.subplot(1, 2, 2)
plt.imshow(image2_rgb)
plt.title('Frame 2')
plt.axis('off')

# Show the plot
plt.show()
```
